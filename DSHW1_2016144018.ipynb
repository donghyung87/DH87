#### ANSWER

```python
words = {1: 'data', 2:'structure', 3:'and', 4:'algorithms', 5: 'pieceofcake'}
```

```python
# CODE HERE
for i in words:
    l=list(words[i])
    t=l[i-1].upper()
    l[i-1]=t
    x=''.join(l)
    words[i]=x
```

### \# PROBLEM IV (50 POINTS)

Complete $\texttt{Complex}$ class below and run some test codes to check that your output is appropriate.   
To this end, you should complete  $\texttt{magnitude}$ and $\texttt{phase}$ functions beforehand.    
Inputs and outputs of these functions are specifed as follows:  
> * Inputs are complex numbers, which can be represented as a list.  
>  For example, $a+bj$, where $j$ is $\sqrt{-1}$, are represented as [a,b].  
> * The $\texttt{magnitude}$ function returns <U>magnitude</U> of a complex number: i.e., $\sqrt{a^2 + b^2}$
> * The  $\texttt{phase}$ function returns <U>phase</U> of a complex number in <U>degree</U>: i.e., $\arctan({{b}\over{a}})$  
>     - To implement the $\arctan$ function in python, import the $\texttt{math}$ library.  
>     - The statement $\texttt{math.atan}(x)$ returns $\arctan(x)$ value in radians.  
>     - Use $\pi$ = 3.141592  
> * Output of each operation should be a $\texttt{Complex}$ class.  

Note that mathematical operations (for example, multiplication) of complex values differ from those of real values.  
> *Also, mathematically, a comparison between two complex numbers using the operators $(<, <=, >, >=)$ is not defined.*  
> *In this problem, we will compare two complex numbers by their magnitude.*  




#### ANSWER

```python
import math
```

```python
def magnitude(complex_number):
    # CODE HERE
    x= math.sqrt(complex_number[0]**2+complex_number[1]**2)
    return x

def phase(complex_number):
    # CODE HERE
    k=math.atan(complex_number[1]/complex_number[0])
    t=k*180/3.141592
    return t

```

```python
class Complex:
    def __init__(self, complex_number):
        # CODE HERE
        self.complex_number = complex_number

    def __str__(self):
        # CODE HERE
        return '{}+{}j'.format(self.complex_number[0],self.complex_number[1])
            
    def __add__(self, other):
        # CODE HERE
        a= self.complex_number[0]+other.complex_number[0]
        b= self.complex_number[1]+other.complex_number[1]
        return '{}+{}j'.format(a,b)
    def __sub__(self, other):
        # CODE HERE
        a= self.complex_number[0]-other.complex_number[0]
        b= self.complex_number[1]-other.complex_number[1]
        if b>=0:
            return '{}+{}j'.format(a,b)
        if b<0:
            return '{}{}j'.format(a,b)
        
    def __mul__(self, other):
        # CODE HERE
        a= self.complex_number[0]*other.complex_number[0]
        b= self.complex_number[1]*other.complex_number[1]
        c=self.complex_number[0]*other.complex_number[1]
        d=self.complex_number[1]*other.complex_number[0]
        return '{}+{}j'.format(a-b,c+d)
    def __eq__(self, other):
        # CODE HERE
        if self.complex_number[0]==other.complex_number[0] and self.complex_number[1]==other.complex_number[1]:
            return True
        else:
            return False
    
    def __ne__(self, other):
        # CODE HERE
        if self.complex_number[0]!=other.complex_number[0] or self.complex_number[1]!=other.complex_number[1]:
            return True
        else:
            return False
    
    def __lt__(self, other):
        # CODE HERE
        if magnitude(self.complex_number) < magnitude(other.complex_number):
            return True
        else:
            return False
    def __le__(self, other):
        # CODE HERE
        if magnitude(self.complex_number) <= magnitude(other.complex_number):
            return True
        else:
            return False
        
    def __gt__(self, other):
        # CODE HERE
        if magnitude(self.complex_number) > magnitude(other.complex_number):
            return True
        else:
            return False
    def __ge__(self, other):
        # CODE HERE
        if magnitude(self.complex_number) >= magnitude(other.complex_number):
            return True
        else:
            return False
```

```python
c1 = [1, 3**(0.5)]
print(magnitude(c1)) #output = 2
print(phase(c1)) #output = 60

c2 = Complex([1, 3])
c3 = Complex([2, 5])
print('add: ', c2 + c3)
print('sub: ', c2 - c3)
print('mul: ', c2*c3)
print('eq: ', c2 == c3)
print('ne: ', c2 != c3)
print('ge: ', c2 >= c3)
print('gt: ', c2 > c3)
print('le: ', c2 <= c3)
print('lt: ', c2 < c3)
```
